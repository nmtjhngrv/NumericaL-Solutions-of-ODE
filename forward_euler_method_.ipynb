Consider 


y'=-100*(y-sint) tâ‰¥0  y(0)=1.

Apply Forward Euler Method with h=0.1 to approximate y(2)? 


Write a program to obtain it. Graph the solution

```python
import pandas as pd
import matplotlib.pyplot as plt
import math
def y__(t,y):
    return -100*(y-math.sin(t)) #function y'=-100*(y-sint)

def yn(h,n):  
    y=[1]
    t=[0]
    m=0
    number_iteration = int(n/h) 
    for i in range(number_iteration):
        y_temp = y__(t[i],y[i])
        y.append(h*y_temp+y[i])
        t.append(m+h)
        m=m +h
    return y

a=yn(0.1,2)
plt.figure(figsize=(6,6))
plt.title("h=0.1 Iteration")
plt.xlabel("Number of Iteration")
plt.ylabel("Number of Y")
plt.plot(a)
plt.show()
```

```python
a=yn(0.01,2)
plt.figure(figsize=(6,6))
plt.plot(a)
plt.title("h=0.01 Iteration")
plt.xlabel("Number of Iteration")
plt.ylabel("Number of Y")
plt.show()
```
